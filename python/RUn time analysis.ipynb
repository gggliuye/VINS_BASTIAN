```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def filter_outlier(data, threshold = 0.7):
    average = np.mean(data)
    derivation = np.abs(data - average)
    max_deri = np.max(derivation, axis=0)
    flag = derivation < max_deri*threshold
    return data[flag]

def plot_analysis_one(data, title):
    data = filter_outlier(data)
    average = np.mean(data)
    line_average = average * np.ones(data.shape);
    plt.plot(data); 
    plt.plot(line_average, '--'); 
    plt.legend(['raw_data', 'average: '+str(average)[0:5]])
    plt.title(title)

def plot_runtime(run_time):
    
    plt.figure(figsize=(15,5))
    plt.subplot(121)
    plot_analysis_one(run_time[:,0], "Solve Problem")
    plt.subplot(122)
    plot_analysis_one(run_time[:,1], "Build Hessian")
```

# Raw method

* window size 10, using direct method for solving linear system, using LM update strategy.

```python
raw_txt = np.loadtxt("../bin/time_D_w10.txt")
print("Read", raw_txt.shape[0], "lines.")
plot_runtime(raw_txt)
```
